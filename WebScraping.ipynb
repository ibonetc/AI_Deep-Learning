## Beautiful Soup para analizar el código HTML

Beautiful Soup es una biblioteca de Python para analizar datos estructurados.

```python
!pip install beautifulsoup4
```

```python
from bs4 import BeautifulSoup
```

Vamos a crear un objeto Beautiful Soup que toma *page.content*, que es el contenido HTML que extrajo anteriormente, como su entrada.

Se usa *page.content* en lugar de *page.text* para evitar problemas con la codificación de caracteres. El atributo *.content* contiene bytes sin procesar, que se pueden decodificar mejor que la representación de texto que imprimió anteriormente con el atributo *.text*.

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
soup
```

## Búsqueda de elementos en una página HTML

En HTML cada elemento puede tener un ID o class asignado.
Por ejemplo podemos explorar en HTML los elementos que contienen las ofertas de trabajo, qué clase tienen.

```python
results = soup.find_all("div", class_="result-item")
```

```python
results[0]
```

```python
for job_element in results:
  title_element = job_element.find("h2", class_="h4 item-title ee-mod")
  company_element = job_element.find("span", class_="info-company-name js-offer-company")
  location_element = job_element.find("span", class_="info-city js-offer-city")
  print(title_element)
  print(company_element)
  print(location_element)
```

## Extraer el texto de los elementos

```python
for job_element in results:
  title_element = job_element.find("h2", class_="h4 item-title ee-mod")
  company_element = job_element.find("span", class_="info-company-name js-offer-company")
  location_element = job_element.find("span", class_="info-city js-offer-city")
  print(title_element.text)
  print(company_element.text)
  print(location_element.text)
```

```python
for job_element in results:
  title_element = job_element.find("h2", class_="h4 item-title ee-mod")
  company_element = job_element.find("span", class_="info-company-name js-offer-company")
  location_element = job_element.find("span", class_="info-city js-offer-city")
  print(title_element.text.strip())
  print(company_element.text.strip())
  print(location_element.text.strip())
  print()
```

También podemos hacer búsqueda por palabras dentro de algunos elementos.

Por ejemplo, sabiendo que en h2 están los títulos de los trabajos encontrados, podemos filtrar y escoger los que tengan "inteligencia artificial" en el título

## Extraer links de la página

```python
for job_element in results:
  links = job_element.find_all("a")
  for link in links:
    link_url = link["href"]
    print(f"Aplicar: {link_url}\n")
```

## Otros paquetes

Un paquete útil para web scraping que la biblioteca estándar de Python ***urllib***, que contiene herramientas para trabajar con URL.

En particular, el módulo **urllib.request** contiene una función llamada **urlopen()** que se puede usar para abrir una URL dentro de un programa.

```python
from urllib.request import urlopen
```

```python
page = urlopen(url)
page
```

```python
html_bytes = page.read()
html = html_bytes.decode("utf-8")
print(html)
```
